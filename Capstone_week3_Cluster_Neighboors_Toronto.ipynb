## 3rd Part
### Cluster and Map

Create new dataframe that only contains Toronto as Borough & drop the extra index column

```python
toronto_data=merged_frame[merged_frame['Borough'].str.contains("Toronto")].reset_index()
df_toronto=toronto_data.drop('index',1)
```

5 clusters will be made based on the locations(latitude and longitude) values. In order to do clustering non numerical columns are dropped.


```python
# set number of clusters
kclusters = 5

toronto_clus = df_toronto.drop('Neighbourhood', 1)
toronto_cluster = toronto_clus.drop('Borough', 1)
toronto_clustering = toronto_cluster.drop('Postal Code', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

Add the labels list as a column to dataframe 

```python
toronto_clustering['labels']=kmeans.labels_
toronto_clustering
```

Now the clustering is done, in order to have more comprehensive table we merge 2 dataframes, in result there are 2 longitude columns, so I have dropped one and change the name of other.

```python
toronto_merged = pd.merge(df_toronto, toronto_clustering , on='Latitude', how='inner')
toronto_merged=toronto_merged.rename(columns={"Longitude_x": "Longitude"})
toronto_merged=toronto_merged.drop('Longitude_y',1)
```

Creating the map

```python

# create map
map_clusters = folium.Map(location=[43.654260, -79.360636], zoom_start=11)

# set color scheme for the clusters
x = np.arange(kclusters)
ys = [i + x + (i*x)**2 for i in range(kclusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, poi, cluster in zip(toronto_merged['Latitude'], toronto_merged['Longitude'], toronto_merged['Neighbourhood'], toronto_merged['labels']):
    label = folium.Popup(str(poi) + ' Cluster ' + str(cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.7).add_to(map_clusters)
       
map_clusters
```
